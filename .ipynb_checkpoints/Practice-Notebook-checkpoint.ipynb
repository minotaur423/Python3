```python
# Problem 5
def min_three(a,b,c):
    if a < b and a < c:
        return a
    elif c < a and c < b:
        return c
    else:
        return b
    
nums = [(7, 3, 5), (15, 20, 40), (300, 550, 137)]
for a,b,c in nums:
    print(min_three(a,b,c),end=" ")
```

```python
# Problem 22
def two_prnt(x,y,n):
    time = 0
    pages = 0
    f_time = 0
    s_time = 0
    faster = 0
    slower = 0
    max_time = 0
    if x < y:
        faster = x
        slower = y
        max_time = slower * n
    elif y < x:
        faster = y
        slower = x
        max_time = slower * n
    else:
        slower = x
        faster = y
        max_time = (slower + 1) * n
    for i in range(1, max_time):
        if pages >= n:
            time = max(f_time,s_time)
            break
        else:
            if i % faster == 0:
                pages += 1
                f_time += faster
            if i % slower == 0:
                pages += 1
                s_time += slower
    return time

grps = ([105684995, 96944250, 8], [12, 88, 10647638], [1, 1, 927175744], [4979224, 7901223, 94], [388996, 376949, 772], [90334, 23099, 5050], [705, 2744, 294119], [68, 3504, 215562], [6, 8, 85993693], [1, 1, 658189901], [760, 995, 749854], [1288, 2455, 379320], [3, 2, 78874497], [407746, 1383237, 680], [2, 2, 484393235], [52, 73, 7621777])
for a,b,c in grps:
    print(two_prnt(a,b,c), end=" ")
```

```python
# Problem 23
def array_chksm(args):
    result = 0
    value = 113
    modulo = 10000007
    for num in range(len(args)):
        result = (((args[num] + result) * value) % modulo)
    return (result)
 
data = [9536, 97, 43912, 800, 79, 8, 1875, 68, 80, 1, 817, 2502, 15749, 32, 2, 7, 6, 1226, 23801, 21770, 0, 68509, 24, 52, 4, 5584, 4, 12045, 115, 565, 61, 43, 10, 151, 5, 17, 354, 75705, 220, 63, 901, 8, 100, 80182, 3, 41, 62195, -1]
swaps = 0
for i in range(len(data)-1):
    if data[i] > data[i+1] and data[i+1] != -1:
        data[i], data[i+1] = data[i+1], data[i]
        swaps += 1
    else:
        continue
data.pop()
print(swaps, array_chksm(data), end=" ")
```

```python
# Problem 24
def randgen(num):
    iterations = 1
    num_list = []
    squared_num = 0
    new_num = num
    while True:
        num_list.append(new_num)
        squared_num = new_num**2
        new_num = int((squared_num / 100) % 10000)
        if new_num in num_list:
            break
        else:
            iterations += 1
    return iterations

for i in (1, 4100, 5761):
    print(randgen(i), end=" ")
```
