```python
# Function Overloading Demo
class ServiceInfo(object):
    def __init__(self, id, name, description):
        self.id = id
        self.name = name
        self.description = description

    # Default arguments for omitting calling arguments
    def alter(self, id, name=None, description=None):
        self.id = id
        if name != None:
            self.name = name
        if description != None:
            self.description = description

    # String representation for printing
    def __str__(self):
        return f'{self.id} {self.name}: {self.description}'

s = ServiceInfo(1021, 'synchronization-job', 'File synchronization process.')
print(s)
s.alter(1021, 'sync-job', 'File sync.')
print(s)
s.alter(1050)
print(s)
```

```python
# Inheritance demo
class Process(object):
    def __init__(self):
        print('init Process')

    def process_function(self):
        return "process_function"

class Client(Process):
    def __init__ (self):
        print('init Client')

    def process_function(self):
        return "client_function"

class Server(Process):
    def __init__(self):
        print('init Server')

    def process_function(self):
        result = super(Server,self).process_function()
        return result + ", server_function"

class BackgroundProcess(Process):
    pass

p = Process()
c = Client()
s = Server()
b = BackgroundProcess()

processes = [p, c, s, b]

print()
for m in processes:
    print(m.process_function())
```

```python
# Print the longest word in a phrase program
def LongestWord(sen):
    sen1 = sen.split()
    length = 0
    for word in sen1:
        if len(word) > length:
            length = len(word)
            longestwrd = word
    return longestwrd

print(LongestWord("This is a testing1234 of this function"))
```

```python
# Operator Overloading Demo
"""
https://docs.python.org/3/reference/datamodel.html
object.__lt__(self, other)
object.__le__(self, other)
object.__eq__(self, other)
object.__ne__(self, other)
object.__gt__(self, other)
object.__ge__(self, other)

object.__add__(self, other)
object.__sub__(self, other)
object.__mul__(self, other)
object.__matmul__(self, other)
object.__truediv__(self, other)
object.__floordiv__(self, other)
object.__mod__(self, other)
object.__divmod__(self, other)
object.__pow__(self, other[, modulo])
object.__lshift__(self, other)
object.__rshift__(self, other)
object.__and__(self, other)
object.__xor__(self, other)
object.__or__(self, other)
"""
class Properties(object):

    def __init__(self, props):
        self.props = []
        if type(props) in (list,tuple):
            self.props.extend(props)

    def __add__(self, other):
        return Properties(self.props + other.props)

    def __sub__(self, other):
        return Properties([p for p in self.props if p not in other.props])

    def __str__(self):
        return f'Properties: {self.props}'

p1 = Properties(['red', 'round', 'small'])
p2 = Properties(['shiny', 'reflective', 'transparent'])
p3 = Properties(['small', 'round'])

print('1. p1: ' + str(p1))
print('2. p2: ' + str(p2))
print('3. p3: ' + str(p3))

print()

p4 = p1 + p2
print('4. p4: ' + str(p4))

print()

print('5. p4 - p3: ' + str(p4 - p3))
```

```python
# The range function generates a sequence of integers
a_range = range(5)
print('1. a_range ->', a_range)
print('2. list(a_range) ->', list(a_range))
# It is often used to execute a "for" loop a number of times
print('3.', end=' ')
for i in range(5):
    print(i, end=' ') # executed five times
print()
# It is similar to the slice function with a start, stop and step
a_range = range(10) # stop only
print('4. list(a_range) ->', list(a_range))
a_range = range(10, 16) # start and stop
print('5. list(a_range) ->', list(a_range))
a_range = range(10, -1, -1) # start, stop and step
print('6. list(a_range) ->', list(a_range))
```

```python
# The set class provides a mapping of unique immutable elements
empty_set = set()
print('1. empty_set ->', empty_set)
alpha = set(('a', 'b', 'c', 'd'))
print('2. alpha ->', alpha)
dup_list = ['c', 'd', 'c', 'd', 'e', 'f']
beta = set(dup_list)
print('3. beta ->', beta)
uniq_list = list(beta)
print('4. uniq_list ->', uniq_list)

gamma = alpha.union(beta)
print('5. gamma ->', gamma)
gamma = alpha | beta
print('6. gamma ->', gamma)

delta = alpha.intersection(beta)
print('7. delta ->', delta)
delta = alpha & beta
print('8. delta ->', delta)

epsilon = alpha.difference(beta)
print('9. epsilon ->', epsilon)
epsilon = alpha - beta
print('10. epsilon ->', epsilon)

eta = alpha.symmetric_difference(beta)
print('11. eta ->', eta)
eta = alpha ^ beta
print('12. eta ->', eta)

print('13. epsilon.isdisjoint(delta) ->', epsilon.isdisjoint(delta))
print('14. epsilon.isdisjoint(eta) ->', epsilon.isdisjoint(eta))

print('15. epsilon.issubset(eta) ->', epsilon.issubset(eta))
print('16. epsilon.issubset(beta) ->', epsilon.issubset(beta))

print('17. eta.issuperset(epsilon) ->', eta.issuperset(epsilon))
print('18. beta.issuperset(epsilon) ->', beta.issuperset(epsilon))

feta = frozenset(eta) # frozensets are immutable without updating methods
print('19. feta ->', feta)

zeta = set()
print('20. zeta ->', zeta)
zeta.add(3)
print('21. zeta ->', zeta)
zeta.add(3)
print('22. zeta ->', zeta)
zeta.add(4)
print('23. zeta ->', zeta)

print('24. gamma ->', gamma)
gamma.discard('a')
print('25. gamma ->', gamma)
gamma.discard('z')
print('26. gamma ->', gamma)
gamma.remove('b')
print('27. gamma ->', gamma)
random_element = gamma.pop()
print('28. random_element ->', random_element)
print('29. gamma ->', gamma)

zeta_ref = zeta
zeta_copy = zeta.copy()
zeta.clear()
print('30. zeta ->', zeta)
print('31. zeta_ref ->', zeta_ref)
print('32. zeta_copy ->', zeta_copy)

print('33. alpha ->', alpha)
alpha_diff = alpha.copy()
alpha_diff.difference_update(beta)
print('34. alpha_diff ->', alpha_diff)

alpha_intersect = alpha.copy()
alpha_intersect.intersection_update(beta)
print('35. alpha_intersect ->', alpha_intersect)

alpha_sym_diff = alpha.copy()
alpha_sym_diff.symmetric_difference_update(beta)
print('36. alpha_sym_diff ->', alpha_sym_diff)

alpha_union = alpha.copy()
alpha_union.update(beta)
print('37. alpha_union ->', alpha_union)
```

```python
# Slicing allows access one or more elements of a sequence
# Immutable sequences include tuples, strings, and bytes
a_tuple = ('a', 1, 2, (3, 4))
a_string = 'immutable'
a_bytes = b'sequence'
# Mutable sequences include lists and bytearrays
a_list = [5, 6, 7, 8, (4, 5)]
a_byte_array = bytearray(b'mutable')
# Accessing is allowed in all sequences
print('1. a_tuple[0] ->', a_tuple[0])
print('2. a_string[1] ->', a_string[1])
print('3. a_bytes[2] ->', a_bytes[2])
print('4. a_list[3] ->', a_list[3])
print('5. a_byte_array[4] ->', a_byte_array[4])
# Negative indexes are from the end
print('6. a_tuple[-1] ->', a_tuple[-1])
print('7. a_string[-2] ->', a_string[-2])
print('8. a_bytes[-3] ->', a_bytes[-3])
print('9. a_list[-4] ->', a_list[-4])
print('10. a_byte_array[-5] ->', a_byte_array[-5])
# Subslices can be accessed with two indexes
print('11. a_list[0:2] ->', a_list[0:2])
print('12. a_list[:2] ->', a_list[:2])
print('13. a_list[2:5] ->', a_list[2:5])
print('14. a_list[2:] ->', a_list[2:])
print('15. a_list[:] ->', a_list[:])
list_ref = a_list
print('16. a_list is list_ref ->', a_list is list_ref)
list_copy = a_list[:]
print('17. a_list is list_copy ->', a_list is list_copy)
# Steps can be taken with a third parameter:
print('18. a_list[::2] ->', a_list[::2])
print('19. a_list[1:4:2] ->', a_list[1:4:2])
print('20. a_string[::-1] ->', a_string[::-1])
# Use additional slices to access elements with sequences
print('21. a_list ->', a_list)
print('22. a_list[4] ->', a_list[4])
print('23. a_list[4][0] ->', a_list[4][0])
print('24. a_list[4][1] ->', a_list[4][1])
# Mutable sequences can be updated with slices
print('25. a_list ->', a_list)
a_list[0] = 'five'
print('26. a_list ->', a_list)
a_list[1:4] = [10, 11, 12]
print('27. a_list ->', a_list)
# A slice object can be used in the [ ] for slicing
a_slice = slice(4)
print('28. a_slice ->', a_slice)
print('29. a_list[a_slice] ->', a_list[a_slice])
a_slice = slice(1,5)
print('30. a_slice ->', a_slice)
print('31. a_list[a_slice] ->', a_list[a_slice])
a_slice = slice(1,5,2)
print('32. a_slice ->', a_slice)
print('33. a_list[a_slice] ->', a_list[a_slice])
```
